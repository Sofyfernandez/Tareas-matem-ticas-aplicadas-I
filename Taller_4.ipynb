### 1. Serie de fibonacci

Escribe un programa en Python que genere los primeros n números
de la serie de Fibonacci y visualízalos en un gráfico de líneas utilizando Matplotlib. El usuario debe poder elegir el número n


```python
import matplotlib.pyplot as plt

# Genera los primeros n números de la serie de Fibonacci
def fibonacci(n):
    fib_series = [0, 1]
    for i in range(2, n):
        fib_series.append(fib_series[-1] + fib_series[-2])
    return fib_series[:n]


n = int(input("Ingresa la cantidad de números de la serie Fibonacci que deseas visualizar: "))

serie = fibonacci(n)

plt.plot(range(n), serie, marker='o', color='r', label="Fibonacci")

# Etiquetas y título
plt.xlabel("Índice")
plt.ylabel("Valor de Fibonacci")
plt.title("Serie de Fibonacci")
plt.legend()
plt.grid()

# Mostrar el gráfico
plt.show()
```

### 2. Análisis de datos meteorológicos

Análisis de datos meteorológicos: Dado un conjunto de datos con las temperaturas diarias de una ciudad durante un año, escribe un programa que calcule la media, la mediana y la desviación estándar de las temperaturas. Luego, crea un gráfico de líneas que muestre la temperatura media mensual

temperaturas = [25.6, 26.1, 24.8, 27.3, 28.4, 26.7, 25.9, 29.1, 30.2, 27.8, 24.3, 23.9, 26.5, 28.0, 27.4, 29.3, 30.5, 26.8, 25.7, 24.6]

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulación de temperaturas diarias para 365 días
temperaturas = np.random.uniform(15, 35, 365)  # Valores entre 15°C y 35°C

# Dividir temperaturas en 12 meses
meses = np.array_split(temperaturas, 12)

# Calcular la media mensual
media_mensual = [np.mean(mes) for mes in meses]

# Cálculos estadísticos
media = np.mean(temperaturas)
mediana = np.median(temperaturas)
desviacion_estandar = np.std(temperaturas)

print(f"Media anual: {media:.2f}°C")
print(f"Mediana anual: {mediana:.2f}°C")
print(f"Desviación estándar: {desviacion_estandar:.2f}°C")

# Graficar temperatura media mensual
plt.figure(figsize=(10, 5))
plt.plot(range(1, 13), media_mensual, marker='o', linestyle='-', color='b', label="Temp. Media Mensual")
plt.axhline(media, color='r', linestyle='dashed', label="Media Anual")

plt.xlabel("Mes")
plt.ylabel("Temperatura (°C)")
plt.title("Temperatura Media Mensual")
plt.xticks(range(1, 13), ["Ene", "Feb", "Mar", "Abr", "May", "Jun", "Jul", "Ago", "Sep", "Oct", "Nov", "Dic"])
plt.legend()
plt.grid()
plt.show()
```

### 4. Gráfica de funciones trigonométricas

Escribe un programa que grafique las
funciones seno, coseno y tangente en el mismo plano para valores entre 0 y 2π

### 6. Pendiente de una recta

Escribe un programa que tome dos puntos en el plano (x1, y1)
y (x2, y2), calcule la pendiente de la recta que los une, y grafique dicha recta

```python
import numpy as np
import matplotlib.pyplot as plt

def graficar_recta(p1, p2):
    x1, y1 = p1
    x2, y2 = p2

    if x1 == x2:  # Recta vertical
        x = np.full(100, x1)
        y = np.linspace(min(y1, y2), max(y1, y2), 100)
    else:
        m = (y2 - y1)/(x2 - x1)
        x = np.linspace(x1-2, x2+2, 100)
        y = m*(x - x1) + y1

    plt.scatter([x1, x2], [y1, y2], c='red')
    plt.plot(x, y)
    plt.title(f"Pendiente: {m:.2f}" if x1 != x2 else "Recta vertical")
    plt.grid(True)
    plt.show()

# Ejemplo:
graficar_recta((2, 3), (5, 8))
```

### 7. Movimiento parabólico

Movimiento parabólico: Escribe un programa que simule el movimiento de un
proyectil bajo la influencia de la gravedad (sin resistencia del aire). Grafica la trayectoria parabólica para diferentes ángulos de lanzamiento



```python
import numpy as np
import matplotlib.pyplot as plt

velocidad_inicial = 20
angulos_lanzamiento = [30, 45, 60]
g = 9.81
dt = 0.01

plt.figure(figsize=(10, 5))

for angulo in angulos_lanzamiento:
    theta = np.radians(angulo)
    v0x = velocidad_inicial * np.cos(theta)
    v0y = velocidad_inicial * np.sin(theta)

    t_total = (2 * v0y) / g
    t = np.arange(0, t_total, dt)

    x = v0x * t
    y = v0y * t - 0.5 * g * t**2

    plt.plot(x, y, label=f'{angulo}°')

plt.xlabel('Distancia (m)')
plt.ylabel('Altura (m)')
plt.title('Movimiento Parabólico de un Proyectil')
plt.legend()
plt.grid()
plt.show()
```

### 8. Deslizador

Construya un formulario con deslizador que modifique la potencia de la f unción $𝑓(𝑥) =𝑥^𝑛$ con n ϵ 𝗡 y n ϵ [1,10]. Grafique todas las funciones



```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

def plot_function(n):
    x = np.linspace(-1, 1, 200)
    plt.figure(figsize=(8, 6))

    # Graficar todas las funciones desde x^1 hasta x^10
    for i in range(1, 11):
        plt.plot(x, x**i, linestyle='dashed' if i != n else 'solid',
                 label=f'$x^{i}$', alpha=0.6 if i != n else 1, linewidth=2 if i == n else 1)

    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.legend()
    plt.title(f'Función $x^{n}$')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

# Crear un deslizador
slider = widgets.IntSlider(value=1, min=1, max=10, step=1, description='Potencia n:')
widgets.interactive(plot_function, n=slider)
```

### 9. (Punto Extra) Fractales con Python:

Escribe un programa que genere el fractal de Sierpinski utilizando ciclos (while), y visualiza el resultado con Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

def generar_triangulo_pascal(n):

    T = np.zeros((n, n), dtype=int)
    T[0, 0] = 1
    fila = 1
    while fila < n:
        T[fila, 0] = 1
        T[fila, fila] = 1
        col = 1

        while col < fila:
            T[fila, col] = T[fila-1, col-1] + T[fila-1, col]
            col += 1
        fila += 1
    return T

def visualizar_fractal(T):

    indices = np.argwhere(T % 2 == 1)

    x = indices[:, 1] - indices[:, 0] / 2.0
    y = -indices[:, 0]

    plt.figure(figsize=(10, 10))
    plt.scatter(x, y, s=10, color='black')
    plt.axis('off')
    plt.show()


tamano = int(input("Digite el tamaño del fractal: "))
triangulo = generar_triangulo_pascal(tamano)
visualizar_fractal(triangulo)
```
