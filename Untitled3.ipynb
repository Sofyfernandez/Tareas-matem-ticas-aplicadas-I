```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

x = np.linspace(-20, 20, 1000)

y = f(x)  # Apply the function to all x values

plt.plot(x, y, 'x')
plt.xlabel("x")
plt.ylabel("f(x) = x^2")
plt.title("Plot of f(x) = x^2")
plt.grid(True)
```
