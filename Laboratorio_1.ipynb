### Calculadroa

```python
import math as mt

def suma(n1, n2):
    return n1 + n2

def resta(n1, n2):
    return n1 - n2

def multiplicar(n1, n2):
    return n1 * n2

def division(n1, n2):
    if n2 == 0:
        return "Error: división por cero."
    return n1 / n2

def raiz_cuadrada(n1):
    if n1 < 0:
        return "Error: no se puede calcular la raíz cuadrada de un número negativo."
    return mt.sqrt(n1)

def elevado_cuadrado(n1):
    return n1 ** 2

def main():
    print("============================")
    print("Bienvenido a mi calculadora")
    print("============================")

    opciones_validas = [0, 1, 2, 3, 4, 5, 6]

    while True:
        try:
            opcion = int(input("""
            ¿Qué operación deseas realizar?
                1. Suma
                2. Resta
                3. Multiplicación
                4. División
                5. Raíz cuadrada
                6. Elevado al cuadrado
                0. Salir
            Elige una opción: """))

            if opcion not in opciones_validas:
                print("Selecciona una opción válida.\n")
                continue

            if opcion == 0:
                print("Saliendo de la calculadora. ¡Hasta luego!")
                break

            # Operaciones que requieren dos números
            if opcion in [1, 2, 3, 4]:
                n1 = float(input("Ingresa el primer número: "))
                n2 = float(input("Ingresa el segundo número: "))
                if opcion == 1:
                    resultado = suma(n1, n2)
                elif opcion == 2:
                    resultado = resta(n1, n2)
                elif opcion == 3:
                    resultado = multiplicar(n1, n2)
                elif opcion == 4:
                    resultado = division(n1, n2)

            # Operaciones que requieren un solo número
            elif opcion in [5, 6]:
                n1 = float(input("Ingresa el número: "))
                if opcion == 5:
                    resultado = raiz_cuadrada(n1)
                elif opcion == 6:
                    resultado = elevado_cuadrado(n1)

            print("El resultado es:", resultado, "\n")

        except ValueError:
            print("Error: Digita un número válido.\n")

if __name__ == "__main__":
    main()
```

### Cajero

```python
import numpy as np
import random as rd

usuarios = {}  # Diccionario para guardar usuarios: {nombre: [contraseña, saldo]}

def generar_saldo_inicial():

    saldos_posibles = np.linspace(100, 1000, 10, dtype=int)
    saldo = rd.choice(saldos_posibles)
    return saldo

def valoresRegistro():

    while True:
        print("\n--- Registro ---")
        nombre = input("Nombre: ").strip().lower()
        if not nombre:
            print("El nombre no puede estar vacío")
            continue

        if nombre in usuarios:
            print("Este usuario ya existe")
            continue

        contraseña = input("Contraseña (mínimo 4 caracteres): ").strip()
        if len(contraseña) < 4:
            print("La contraseña debe tener mínimo 4 caracteres")
            continue


        saldo = generar_saldo_inicial()
        print(f"Tu saldo inicial es: ${saldo}")

        # Guardar en el diccionario
        usuarios[nombre] = [contraseña, saldo]
        print("¡Registro exitoso!")
        return nombre

def iniciar_sesion():
    intentos = 3
    while intentos > 0:
        print("\n--- Inicio de Sesión ---")
        nombre = input("Nombre: ").strip().lower()
        contraseña = input("Contraseña: ").strip()

        if nombre in usuarios and usuarios[nombre][0] == contraseña:
            print(f"\n¡Bienvenido {nombre}!")
            return nombre
        else:
            intentos -= 1
            print(f"Credenciales incorrectas. Intentos restantes: {intentos}")

    print("Has agotado tus intentos")
    return None

def operaciones(nombre):
    while True:
        print("\n--- Menú ---")
        print("1. Depositar")
        print("2. Retirar")
        print("3. Consultar saldo")
        print("4. Salir")

        opcion = input("Elige una opción: ")

        if opcion == "1":
            while True:
                try:
                    monto = float(input("Monto a depositar: "))
                    if monto <= 0:
                        print("El monto debe ser positivo")
                    else:
                        usuarios[nombre][1] += monto
                        print(f"Depósito exitoso. Nuevo saldo: {usuarios[nombre][1]}")
                        break
                except ValueError:
                    print("Ingresa un número válido")

        elif opcion == "2":
            while True:
                try:
                    monto = float(input("Monto a retirar: "))
                    if monto <= 0:
                        print("El monto debe ser positivo")
                    elif monto > usuarios[nombre][1]:
                        print("Saldo insuficiente")
                    else:
                        usuarios[nombre][1] -= monto
                        print(f"Retiro exitoso. Nuevo saldo: {usuarios[nombre][1]}")
                        break
                except ValueError:
                    print("Ingresa un número válido")

        elif opcion == "3":
            print(f"Saldo actual: {usuarios[nombre][1]}")

        elif opcion == "4":
            print("¡Hasta pronto!")
            break

        else:
            print("Opción inválida")

def main():
    while True:
        print("\n=== Banco ===")
        print("1. Registrarse")
        print("2. Iniciar sesión")
        print("3. Salir")

        opcion = input("Elige una opción: ")

        if opcion == "1":
            nombre = valoresRegistro()
            if nombre:
                operaciones(nombre)

        elif opcion == "2":
            nombre = iniciar_sesion()
            if nombre:
                operaciones(nombre)

        elif opcion == "3":
            print("Gracias por usar nuestro sistema")
            break

        else:
            print("Opción inválida")

if __name__ == "__main__":
    main()
```
